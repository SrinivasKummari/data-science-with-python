# Import Packages

```python
#import libraries
import numpy as np
import pandas as pd
import os
```

# Import Data

```python
#set the path 
raw_data_path = os.path.join(os.path.pardir, 'data','raw')
train_file_path = os.path.join(raw_data_path, 'train.csv')
test_file_path = os.path.join(raw_data_path, 'test.csv')
```

```python
# read the data with all the default parameters
train_df = pd.read_csv(train_file_path, index_col = 'PassengerId')
test_df = pd.read_csv(test_file_path, index_col = 'PassengerId')
```

```python
#get the type of data frame
type(train_df)
type(test_df)
```

```python
# filter rows based on the condition
passengers_aged_35_or_more = df.loc[df.Age > 35,:]
```

```python
#print
print("Number of passengers whose age greater than 35: {0}".format(len(passengers_aged_35_or_more)))
```

```python
male_passengers = df.loc[df.Sex == 'male', :]
print('Number of male passengers: {0}'.format(len(male_passengers)))
```

```python
male_passengers_greater_35 = df.loc[(df.Sex == 'male')& (df.Age > 35), :]
print('Male Passengers greater 35: {0}'.format(len(male_passengers_greater_35)))
```
